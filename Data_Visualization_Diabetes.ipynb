Relation of high BP with Diabetes_binary

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dm = pd.read_csv("/content/diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
```

```python
labels=["non HighBP","HighBP"]
plt.pie(dm["HighBP"].value_counts() , labels = labels , autopct='%.02f');
```

Checking the relation between highBP and Diabetes_binary

```python
pd.crosstab(dm.HighBP ,dm.Diabetes_binary).plot(kind="bar",figsize=(5,4))

plt.title('Relation between HighBP and Diabetes_binary')
plt.xlabel('HighBP')
plt.ylabel('Frequency')
plt.show()
```

Relation between HighChol and Diabetes_binary

```python
labels = ["non HighChol", "HighChol"]
plt.pie(dm["HighChol"].value_counts(),labels = labels,autopct='%.02f');
```

```python
pd.crosstab(dm.HighChol ,dm.Diabetes_binary).plot(kind="bar",figsize=(5,4))

plt.title('Relation between HighChol and Diabetes_binary')
plt.xlabel('HighChol')
plt.ylabel('Frequency')
plt.show()
```

Checking HighBp and HighChol combined Effect on Diabetes_binary

```python
(dm.groupby(["HighBP","HighChol"])["Diabetes_binary"].value_counts()/dm.groupby(["HighBP","HighChol"])["Diabetes_binary"].count())*100
```

```python
sns.catplot(x=dm['HighBP'],y=dm['HighChol'],hue=dm['Diabetes_binary'],kind="bar");
plt.title("Relation between HighBP,HighChol and Diabetes_binary")
```

```python
plt.figure(figsize=(15,10))
warnings.filterwarnings('ignore')

sns.distplot(dm.BMI[dm.Diabetes_binary == 0], color="green", label="No Diabetes")
sns.distplot(dm.BMI[dm.Diabetes_binary == 1], color="yellow", label="Diabetes")
plt.title("Relation between BMI and Diabetes")

plt.legend()
```
