FEATURE ENGINEERING

```python
#Create new features
#Total stats without HP
df['Total_without_HP']=df['Total'] - df['HP']
```

```python
df.head()
```

```python
#Normalizing features
from sklearn.preprocessing import MinMaxScaler

#Normalize attack ,Defence,speed
scalar= MinMaxScaler()

df[['Attack_scaled','Defence_scaled','Speed_scaled'] ] = scalar.fit_transform(df[['Attack','Defense','Speed']])
```

```python
df.head()
```

```python
#Binning

bins = [0,50,100,150,200]
labels = ['Very Slow','Slow','Fast','Very Fast']

df['Speed_category'] = pd.cut(df['Speed'], bins = bins, labels = labels)
```

```python
fast_pokemons = df[df['Speed_category'] == 'Fast']
print(fast_pokemons)
```

```python
#label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Legendary_Encoded'] = le.fit_transform(df['Legendary'])
```
